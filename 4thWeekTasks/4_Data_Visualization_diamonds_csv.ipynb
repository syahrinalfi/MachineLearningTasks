```python
from google.colab import drive
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Mount Drive
drive.mount('/content/drive')

# Impor File CSV
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Machine Learning/Minggu 4/diamonds.csv", index_col = 0)
```

# Feature Statistic

```python
# Menampilkan Statistik Fitur
df.describe()
```

# Box Plot

Prompt ChatGPT = "Tampilkan informasi data diamonds.csv"

```python
df.info()
```

Prompt ChatGPT = "Tampilkan fitur numerik dalam bentuk boxplot secara terpisah"

```python
# Memilih hanya fitur numerik
numerical_features = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

# Membuat boxplot untuk setiap fitur numerik
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=df[feature])
    plt.title(f'Boxplot of {feature}')
    plt.show()
```

# Scatter Plot

Prompt ChatGPT = "Tampilkan fitur numerik dalam bentuk scatter plot"

```python
# Membuat scatter plot untuk setiap pasangan fitur numerik
sns.set(style="ticks")
sns.pairplot(df[numerical_features])
plt.show()
```

# Distribution Plot

Prompt ChatGPT = "Tampilkan fitur numerik dalam bentuk distribution plot"

```python
# Membuat distribusi plot untuk setiap fitur numerik
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[feature], kde=True, bins=30)
    plt.title(f'Distribution Plot of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```

# Violin Plot

Prompt ChatGPT = "Tampilkan fitur numerik dalam bentuk violin plot"

```python
# Membuat violin plot untuk setiap fitur numerik
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.violinplot(x=df[feature], inner='quartile')
    plt.title(f'Violin Plot of {feature}')
    plt.xlabel(feature)
    plt.show()
```

# Heatmap

Prompt ChatGPT = "Tampilkan heatmap untuk fitur numerik"

```python
# Menghitung korelasi antar fitur numerik
correlation_matrix = df[numerical_features].corr()

# Membuat heatmap untuk korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Heatmap of Numerical Features Correlation')
plt.show()
```
