### Visualisasi Data

Prompt ChatGPT : Beri kode untuk menampilkan pie diagram chart untk fitur Age dengan min 20, max 69 dengan range masing-masing 10 tahun

```python
import matplotlib.pyplot as plt

# Filter data berdasarkan rentang usia
filtered_data = df[(df['Age'] >= 20) & (df['Age'] <= 69)]

# Membuat bins untuk rentang usia masing-masing 10 tahun
bins = [20, 30, 40, 50, 60, 70]
labels = ['20-29', '30-39', '40-49', '50-59', '60-69']

# Mengelompokkan data usia ke dalam bins
age_groups = pd.cut(filtered_data['Age'], bins=bins, labels=labels, right=False)

# Menghitung frekuensi usia dalam setiap grup
age_group_counts = age_groups.value_counts()

# Membuat pie chart
plt.figure(figsize=(8, 8))
plt.pie(age_group_counts, labels=age_group_counts.index, autopct='%1.1f%%')
plt.title('Age Distribution (20-69 years)')
plt.show()
```
